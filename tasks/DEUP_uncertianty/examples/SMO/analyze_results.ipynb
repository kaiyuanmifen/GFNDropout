```python
import pickle
import matplotlib.pyplot as plt
```

```python
gfn = pickle.load(open('pickles/gfnmcdropout_multi_optima_6', 'rb'))['results'].mean(0)
gfn_2 = pickle.load(open('pickles/gfnmcdropout_multi_optima_6_small', 'rb'))['results'].mean(0)
mcdropout = pickle.load(open('pickles/mcdropout_multi_optima_6', 'rb'))['results'].mean(0)
mcdropout_2 = pickle.load(open('pickles/mcdropout_multi_optima_6_small', 'rb'))['results'].mean(0)
deup = pickle.load(open('pickles/deup_multi_optima_6', 'rb'))['results'].mean(0)
gp = pickle.load(open('pickles/gp_multi_optima_6', 'rb'))['results'].mean(0)
```

```python
# HERE, MCDropout and GFNMCDropout (bigmodel) use a 128x128x128 hidden layers neural net
# GFNMCDropout (smallmodel) uses a 32x32 hidden layers neural net
# 5 seeds are used for each method
# All methods are initialized with THE SAME 6 datapoints, for each seed
# The function is multi_optima
# All methods are optimized for 50 steps

plt.figure(figsize=(10, 6))
plt.plot(range(6, 57), gfn, label='GFN-MCDropout (bigmodel)')
plt.plot(range(6, 57), gfn_2, label='GFN-MCDropout (smallmodel)')
plt.plot(range(6, 57), mcdropout, label='MCDropout (bigmodel)')
plt.plot(range(6, 57), mcdropout_2, label='MCDropout (smallmodel)')
plt.plot(range(6, 57), deup, label='DEUP')
plt.plot(range(6, 57), gp, label='GP')
plt.legend()
plt.xlabel('Number of black-box function queries')
plt.ylabel('Best observed value')
plt.show()
```
